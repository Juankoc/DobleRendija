# Simulación del Experimento de la Doble Rendija
Esta simulación genera el patrón de interferencia que se observaría en una pantalla debido al paso de partículas (como electrones o fotones) a través de dos rendijas, destacando el comportamiento ondulatorio de las partículas en la mecánica cuántica.

# Descripción del experimento:
Configuración: Se proyecta una fuente de luz (o partículas) hacia una pantalla con dos rendijas estrechas. Detrás de la pantalla se coloca una segunda pantalla para observar los patrones de interferencia.

Resultados Clásicos: Si se considera a la luz como una partícula (como en la física clásica), se esperaría que las partículas pasen por una u otra rendija y formen dos franjas detrás de las rendijas.

Resultados Cuánticos: Sin embargo, al considerar la luz o partículas como ondas, se produce un patrón de interferencia en la pantalla final. Este patrón está compuesto por zonas brillantes (donde las ondas se refuerzan) y zonas oscuras (donde las ondas se cancelan). Esto sugiere que las partículas actúan como ondas y atraviesan ambas rendijas simultáneamente, interfiriendo consigo mismas.

```python
import numpy as np
import matplotlib.pyplot as plt

largo_onda = 0.5 
distancia_rendijas = 1.0 
distancia_pantalla = 5.0  
num_puntos = 1000  


pantalla = np.linspace(-5, 5, num_puntos)


def intensidad_interferencia(x, largo_onda, distancia_rendijas, distancia_pantalla):

    rendija1 = -distancia_rendijas / 2
    rendija2 = distancia_rendijas / 2
    

    distancia1 = np.sqrt(distancia_pantalla**2 + (x - rendija1)**2)
    distancia2 = np.sqrt(distancia_pantalla**2 + (x - rendija2)**2)
    

    delta_fase = (distancia1 - distancia2) * 2 * np.pi / largo_onda
    

    intensidad = (np.cos(delta_fase / 2))**2
    return intensidad


intensidad = intensidad_interferencia(pantalla, largo_onda, distancia_rendijas, distancia_pantalla)


plt.plot(pantalla, intensidad)
plt.title('Patrón de interferencia en el experimento de la doble rendija')
plt.xlabel('Posición en la pantalla')
plt.ylabel('Intensidad')
plt.show()
```

# Descripción del experimento:
La función intensidad_interferencia calcula la intensidad de la luz en función de la posición en la pantalla. Se basa en la diferencia de fase de las ondas que llegan desde las dos rendijas a cada punto de la pantalla.
Usamos un coseno al cuadrado para modelar la interferencia constructiva y destructiva de las dos ondas.
Finalmente, se grafica el patrón de interferencia resultante, que mostrará las franjas características del experimento.
